```python
####----------- barchart and pie chart----------------------------------------------

import matplotlib.pyplot as plt
import pandas as pd
df=pd.read_csv('/content/student_marks.1.csv')
print(df)
plt.bar(df.student,df.total,edgecolor='black',width=0.4)
plt.xlabel("Students")
plt.ylabel("Total Marks")
plt.title("Student Marks")
plt.xticks(rotation=90)
plt.show()
print()
subject=df[['telugu','hindi','english']].sum()
plt.pie(subject,labels=subject.index,autopct='%1.1f%%',startangle=90,colors=['red','green','blue'])
plt.title('Average Grade')
plt.axis('equal')
plt.legend()
plt.show()
```

```python
###----------grouped bar chart--------------------------------------------------

import matplotlib.pyplot as plt
import numpy as np
platforms=['IMDB','ROTTEN TOMATOES','METACRITIC','book my show']
mahanati=[85,87,96,92]
manam=[80,85,89,92]
kalki=[88,90,79,80]
mahavatar=[99,96,95,98]
rrr=[80,85,90,92]
x=np.arange(len(platforms))
plt.ylim(0,150)
width=0.166
plt.bar(x+(-width-width),mahanati,width,label='mahanati')
plt.bar(x-width,manam,width,label='manam')
plt.bar(x,kalki,width,label='kalki')
plt.bar(x+width,mahavatar,width,label='mahavatar')
plt.bar(x+(width+width),rrr,width,label='rrr')
plt.xticks(x,platforms)
plt.xlabel('Platforms')
plt.ylabel('percentage of ratings')
plt.legend()
plt.title('movies in different platforms')
plt.show()

```
