```python
import matplotlib.pyplot as plt
months=['jan','feb','mar','apr','may','jun']
sales=[3200,4300,5000,7600,6400,8500]
products=[6000,5000,8500,9000,8700,10900]
plt.title('monthly sales')
plt.xlabel('months')
plt.ylabel('sales,products')
plt.grid(True)
plt.plot(months,sales,marker='*',markersize='8',linestyle='--',color='blue')
plt.plot(months,products,marker='*',markersize='8',linestyle='--',color='red')
plt.legend(['sales','products'])
plt.show()
plt.title('monthly sales')
plt.xlabel('months')
plt.ylabel('sales')
plt.bar(months,sales,color=['red','green','blue','yellow','orange','purple'],
         edgecolor='black',label='sales')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
data=pd.read_csv('/content/student_marks.1.csv')
d=pd.DataFrame(data)
e=np.array(d.telugu)
f=np.array(d.english)
g=np.array(d.hindi)
h=np.hstack([e,f,g])
plt.title('subjects marks between')
plt.boxplot(h,notch=False,patch_artist=True,vert=0,
             boxprops={'facecolor':'skyblue'},
             medianprops={'color':'red','linewidth':2},
             whiskerprops={'color':'green','linewidth':2},
             )
plt.show()
```

```python
import matplotlib.pyplot as plt
data=[[57,63,85,23,62,25,82],
       [68,52,52,73,83,32,67],
       [95,58,25,65,57,80,43]
       ]
plt.boxplot(data,notch=False,patch_artist=True,vert=0,
             boxprops={'facecolor':'blue'},
             medianprops={'color':'red','linewidth':2},
             whiskerprops={'color':'green','linewidth':2},)
plt.title('marks')
plt.xlabel('groups')
plt.ylabel('values')
plt.xticks([1,2,3],['Group 1','Group 2','Group 3'])
plt.grid(True)
plt.show()
import matplotlib.pyplot as plt
import seaborn as sns
data=[10,20,25,36,30,48,51,52,50,66,60,74,88,90]
sns.histplot(data, bins=8,kde=True,color='blue')
plt.xlabel('Ages')
plt.ylabel('Frequency')
plt.title('Histogram of Ages')
plt.show()
```
